## Step 5 – Write the Code

Translate your pseudocode into Python code. Below is one example solution:

````python

    """
    Personal Finance Tracker Setup Program
    """

    # Step 1: Start the program
    User_Name = ""
    Email = ""
    Preferred_Currency = ""
    Main_Savings_Goal = ""
    Monthly_Budget = 0.0
    Theme_Color = ""
    Welcome_Message = ""

    # Step 2: Prompt the user for their name and store it
    print("Enter your name:")
    User_Name = input()

    # Step 3: Prompt for their email and store it
    print("Enter your email address:")
    Email = input()

    # Step 4: Prompt for their preferred theme color and store it
    print("Enter your preferred theme color (e.g., Dark Blue, Light Green):")
    Theme_Color = input()

    # Step 5: Prompt for their primary currency and store it
    print("Enter your preferred currency (e.g., USD, EUR, GBP):")
    Preferred_Currency = input()

    # Step 6: Prompt for their main savings goal and store it
    print("Enter your main savings goal:")
    Main_Savings_Goal = input()

    # Step 7: Prompt for their monthly budget and store it
    print("Enter your monthly budget:")
    Monthly_Budget = float(input())

    # Step 8: Construct a welcome and summary message using these inputs
    Welcome_Message = (
        f"Welcome to your Personal Finance Tracker, {User_Name}! \n"
        f"Your profile is ready. The app will use a {Theme_Color} theme "
        f"and all amounts will be in {Preferred_Currency}.\n "
        f"We will send updates to {Email}, "
        f"so Let's start working towards your goal of saving for {Main_Savings_Goal}!"
    )

    # Step 9: Print the summary message
    print(Welcome_Message)

    # Step 10: End the program



````
