### Step 2: Initialize Your Personal Finance Tracker

Create your personalized tracker:

````python
from datetime import datetime

def create_personal_finance_tracker(name="Demo User", monthly_budget=2000):
    """
    Initialize a personalized finance tracker with dictionaries.
    """
    tracker = {
        "metadata": {
            "version": "1.0",
            "created_date": datetime.now().strftime("%Y-%m-%d"),
            "last_updated": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "preferences": {
                "currency": "EUR",
                "date_format": "%d-%m-%Y",
                "budget_cycle": "monthly"
            }
        },

        "user_profile": {
            "name": name,
            "email": f"{name.lower().replace(' ', '.')}@etu.univ.edu",
            "financial_goals": {
                "monthly_budget": monthly_budget,
                "savings_target": monthly_budget * 0.20,   # 20% savings
                "emergency_fund_target": monthly_budget * 3  # 3 months of budget
            }
        },

        "expenses": {},

        "categories": {},

        "analytics": {
            "total_expenses": 0,
            "expense_count": 0,
            "last_expense_id": 0
        },

        "settings": {
            "notifications": {"budget_warnings": True},
            "privacy": {"backup_enabled": True}
        }
    }

    # Define categories with allocations
    category_definitions = {
        "Housing": {"allocation": 0.30, "description": "Rent, utilities", "essential": True},
        "Food": {"allocation": 0.25, "description": "Groceries, meals", "essential": True},
        "Transport": {"allocation": 0.15, "description": "Fuel, transport", "essential": True},
        "Entertainment": {"allocation": 0.10, "description": "Leisure, outings", "essential": False},
        "Savings": {"allocation": 0.20, "description": "Savings, investments", "essential": True}
    }

    # Initialize categories
    for category, config in category_definitions.items():
        budget_amount = monthly_budget * config["allocation"]
        tracker["categories"][category] = {
            "budget_allocation": config["allocation"],
            "budget_amount": budget_amount,
            "actual_spent": 0,
            "remaining_budget": budget_amount,
            "description": config["description"],
            "essential": config["essential"],
            "expense_count": 0
        }

    return tracker

print("🚀 Initializing Finance Tracker")
# User setup
user_name = input("Enter your name: ")

while True:

    try:
        monthly_budget = float(input("Enter your monthly budget: $"))
        print(f"✅ Budget accepted: ${monthly_budget:.2f}")
        my_finance_tracker = create_personal_finance_tracker(user_name, monthly_budget)
        break
    except ValueError:
        print("Please enter a valid number!")

print(f"✅ Tracker created for: {my_finance_tracker['user_profile']['name']}")
print(f"💰 Monthly budget: {my_finance_tracker['user_profile']['financial_goals']['monthly_budget']} EUR")
print(f"🎯 Savings target: {my_finance_tracker['user_profile']['financial_goals']['savings_target']} EUR")
print(f"📅 Created on: {my_finance_tracker['metadata']['created_date']}")

print("\n📊 Initialized categories:")
for category, details in my_finance_tracker['categories'].items():
    percentage = details['budget_allocation'] * 100
    print(f"   {category}: {details['budget_amount']:.2f} EUR ({percentage:.0f}%)")

print("\n🎉 Your finance tracker is ready!")
````

### Step 3: Essential Expense Management Functions

Build the core functions using dictionary best practices:

````python

def generate_unique_expense_id(tracker):
    """
    Generate a unique ID for each expense.
    """
    tracker["analytics"]["last_expense_id"] += 1
    return f"EXP{tracker['analytics']['last_expense_id']:03d}"


def add_expense(tracker, amount, description, category, essential=True, payment_method="Cash", tags=None):
    """
    Add a new expense to the tracker.
    """
    expense_id = generate_unique_expense_id(tracker)
    current_date = datetime.now().strftime("%Y-%m-%d")

    expense = {
        "amount": float(amount),
        "description": description,
        "category": category,
        "date": current_date,
        "payment_method": payment_method,
        "essential": essential,
        "tags": tags if tags else []
    }

    tracker["expenses"][expense_id] = expense

    # Update categories
    if category in tracker["categories"]:
        tracker["categories"][category]["actual_spent"] += amount
        tracker["categories"][category]["remaining_budget"] -= amount
        tracker["categories"][category]["expense_count"] += 1

    # Update analytics
    tracker["analytics"]["total_expenses"] += amount
    tracker["analytics"]["expense_count"] += 1
    tracker["metadata"]["last_updated"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    return expense_id


def get_expense_by_id(tracker, expense_id):
    """
    Retrieve an expense by its ID.
    """
    return tracker["expenses"].get(expense_id, None)


def find_expenses_by_criteria(tracker, **criteria):
    """
    Flexible search for expenses based on multiple criteria.
    """
    results = {}
    for exp_id, exp in tracker["expenses"].items():
        include = True
        if "category" in criteria and exp["category"] != criteria["category"]:
            include = False
        if "essential" in criteria and exp["essential"] != criteria["essential"]:
            include = False
        if "min_amount" in criteria and exp["amount"] < criteria["min_amount"]:
            include = False
        if "max_amount" in criteria and exp["amount"] > criteria["max_amount"]:
            include = False
        if "description_contains" in criteria and criteria["description_contains"].lower() not in exp["description"].lower():
            include = False
        if include:
            results[exp_id] = exp
    return results


def update_expense(tracker, expense_id, **updates):
    """
    Update an existing expense.
    """
    if expense_id not in tracker["expenses"]:
        return False, f"Expense {expense_id} not found"

    expense = tracker["expenses"][expense_id]
    old_amount = expense["amount"]
    old_category = expense["category"]

    for field, value in updates.items():
        if field in expense:
            expense[field] = value

    # If amount or category changes
    if "amount" in updates or "category" in updates:
        new_amount = expense["amount"]
        new_category = expense["category"]

        # Adjust old category
        if old_category in tracker["categories"]:
            tracker["categories"][old_category]["actual_spent"] -= old_amount
            tracker["categories"][old_category]["remaining_budget"] += old_amount
            tracker["categories"][old_category]["expense_count"] -= 1

        # Adjust new category
        if new_category in tracker["categories"]:
            tracker["categories"][new_category]["actual_spent"] += new_amount
            tracker["categories"][new_category]["remaining_budget"] -= new_amount
            tracker["categories"][new_category]["expense_count"] += 1

        # Update totals
        tracker["analytics"]["total_expenses"] = tracker["analytics"]["total_expenses"] - old_amount + new_amount

    tracker["metadata"]["last_updated"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    return True, f"Expense {expense_id} updated successfully"


def delete_expense(tracker, expense_id):
    """
    Delete an expense and update data.
    """
    if expense_id not in tracker["expenses"]:
        return False, f"Expense {expense_id} not found"

    expense = tracker["expenses"][expense_id]
    category = expense["category"]
    amount = expense["amount"]

    if category in tracker["categories"]:
        tracker["categories"][category]["actual_spent"] -= amount
        tracker["categories"][category]["remaining_budget"] += amount
        tracker["categories"][category]["expense_count"] -= 1

    tracker["analytics"]["total_expenses"] -= amount
    tracker["analytics"]["expense_count"] -= 1

    del tracker["expenses"][expense_id]
    tracker["metadata"]["last_updated"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    return True, f"Expense {expense_id} deleted successfully"


# Add expenses
exp1 = add_expense(my_finance_tracker, 1200, "Monthly rent", "Housing", True, "Bank transfer", ["rent"])
exp2 = add_expense(my_finance_tracker, 200, "Weekly groceries", "Food", True, "Debit card", ["groceries"])

print("Total expenses:", my_finance_tracker["analytics"]["total_expenses"])
print("Number of expenses:", my_finance_tracker["analytics"]["expense_count"])

# Search
food_expenses = find_expenses_by_criteria(my_finance_tracker, category="Food")
print("Food expenses:", food_expenses)

# Update
update_expense(my_finance_tracker, exp2, amount=250)

# Delete
delete_expense(my_finance_tracker, exp1)

````

### Step 4: Advanced Analytics and Reporting

Create sophisticated analysis functions:

````python
def calculate_budget_performance(tracker):
    """
    Analyze budget performance by category and overall.
    """
    performance = {}
    monthly_budget = tracker["user_profile"]["financial_goals"]["monthly_budget"]
    total_spent = tracker["analytics"]["total_expenses"]

    # Overall performance
    performance["overall"] = {
        "budget": monthly_budget,
        "spent": total_spent,
        "remaining": monthly_budget - total_spent,
        "percentage_used": (total_spent / monthly_budget * 100) if monthly_budget > 0 else 0,
        "status": "over_budget" if total_spent > monthly_budget else "within_budget"
    }

    # Category performance
    performance["categories"] = {}
    for category, details in tracker["categories"].items():
        spent = details["actual_spent"]
        budget = details["budget_amount"]
        remaining = budget - spent
        percentage = (spent / budget * 100) if budget > 0 else 0

        performance["categories"][category] = {
            "budget": budget,
            "spent": spent,
            "remaining": remaining,
            "percentage_used": percentage,
            "expense_count": details["expense_count"],
            "average_expense": spent / details["expense_count"] if details["expense_count"] > 0 else 0,
            "status": "over_budget" if spent > budget else "within_budget",
            "variance": spent - budget
        }

    return performance


def generate_spending_insights(tracker):
    """
    Generate smart recommendations based on expenses.
    """
    insights = []
    performance = calculate_budget_performance(tracker)
    overall = performance["overall"]

    # Overall insights
    if overall["percentage_used"] >= 90:
        insights.append({"type": "warning", "message": f"🚨 Already {overall['percentage_used']:.1f}% of the budget used"})
    elif overall["percentage_used"] >= 75:
        insights.append({"type": "caution", "message": f"⚡ {overall['percentage_used']:.1f}% of the budget used"})
    elif overall["percentage_used"] <= 50:
        insights.append({"type": "positive", "message": f"🌟 Good management! Only {overall['percentage_used']:.1f}% used"})

    # Category insights
    for category, perf in performance["categories"].items():
        if perf["status"] == "over_budget":
            insights.append({"type": "warning", "message": f"📊 {category} exceeds the budget by {abs(perf['variance']):.2f} EUR"})
        elif perf["percentage_used"] >= 80:
            insights.append({"type": "caution", "message": f"⚠️ {category} is at {perf['percentage_used']:.1f}% of the budget"})
        elif perf["expense_count"] == 0:
            insights.append({"type": "info", "message": f"💡 No expenses recorded for {category}"})

    return insights


def create_comprehensive_report(tracker):
    """
    Generate a complete financial report.
    """
    print("📊 COMPLETE FINANCIAL REPORT")
    print("=" * 60)

    user_name = tracker["user_profile"]["name"]
    report_date = datetime.now().strftime("%Y-%m-%d %H:%M")

    print(f"👤 User: {user_name}")
    print(f"📅 Report generated: {report_date}")
    print(f"💾 Last updated: {tracker['metadata']['last_updated']}")
    print()

    # Overall summary
    performance = calculate_budget_performance(tracker)
    overall = performance["overall"]

    print("💰 Budget summary")
    print("-" * 40)
    print(f"Monthly budget: {overall['budget']:.2f} EUR")
    print(f"Total expenses: {overall['spent']:.2f} EUR")
    print(f"Remaining: {overall['remaining']:.2f} EUR")
    print(f"Usage: {overall['percentage_used']:.1f}%")
    print()

    # Categories
    print("📊 Category details")
    print("-" * 40)
    for category, perf in performance["categories"].items():
        print(f"{category}: {perf['spent']:.2f}/{perf['budget']:.2f} EUR ({perf['percentage_used']:.1f}%)")
    print()

    # Insights
    insights = generate_spending_insights(tracker)
    print("💡 Recommendations")
    print("-" * 40)
    for insight in insights:
        print(f"- {insight['message']}")

    print("=" * 60)


# Add some expenses
add_expense(my_finance_tracker, 1200, "Rent", "Housing", True, "Bank transfer")
add_expense(my_finance_tracker, 200, "Groceries", "Food", True, "Debit card")
add_expense(my_finance_tracker, 80, "Cinema", "Entertainment", False, "Cash")

# Generate a report
create_comprehensive_report(my_finance_tracker)
````

## 🎯 Phase 3: Advanced Features and Interactivity (30 minutes)

### Step 5: Interactive User Interface

Create a user-friendly interface showcasing dictionary capabilities:

````python

def display_main_menu():
    """Display the main menu"""
    print("\n🎛️ MAIN MENU - FINANCE TRACKER")
    print("=" * 40)
    print("1. ➕ Add an expense")
    print("2. 🔍 Search expenses")
    print("3. 📊 Budget summary")
    print("4. 💡 Get financial insights")
    print("5. 📈 Generate a full report")
    print("6. ✏️ Edit an expense")
    print("7. 🗑️ Delete an expense")
    print("0. 🚪 Exit")
    print("=" * 40)


def get_user_input_safe(prompt, input_type=str, options=None):
    """Secure user input with validation"""
    while True:
        try:
            user_input = input(prompt).strip()
            if input_type == float:
                value = float(user_input)
                if value < 0:
                    print("❌ Please enter a positive number")
                    continue
                return value
            elif input_type == int:
                return int(user_input)
            elif options and user_input not in options:
                print(f"❌ Choose from: {', '.join(options)}")
                continue
            else:
                return user_input
        except ValueError:
            print(f"❌ Invalid input. Expected: {input_type.__name__}")


def add_expense_interactive(tracker):
    """Interactive expense addition"""
    print("\n➕ ADD AN EXPENSE")
    amount = get_user_input_safe("Amount (€): ", float)
    description = get_user_input_safe("Description: ")

    categories = list(tracker["categories"].keys())
    print(f"Available categories: {', '.join(categories)}")
    category = get_user_input_safe("Category: ", options=categories)

    essential = get_user_input_safe("Essential? (y/n): ", options=['y', 'n']) == 'y'
    payment_methods = ["Cash", "Card", "Transfer"]
    print(f"Payment methods: {', '.join(payment_methods)}")
    payment_method = get_user_input_safe("Method: ", options=payment_methods)

    expense_id = add_expense(tracker, amount, description, category, essential, payment_method)
    print(f"✅ Expense added with ID {expense_id}")


def run_interactive_finance_tracker(tracker):
    """Main interactive loop"""
    print("🚀 Welcome to your Finance Tracker")
    print(f"👤 User: {tracker['user_profile']['name']}")
    print(f"💰 Monthly budget: {tracker['user_profile']['financial_goals']['monthly_budget']} EUR")

    while True:
        display_main_menu()
        choice = get_user_input_safe("Choice (0-7): ", options=[str(i) for i in range(8)])

        if choice == "1":
            add_expense_interactive(tracker)
        elif choice == "2":
            keyword = get_user_input_safe("Keyword: ")
            results = find_expenses_by_criteria(tracker, description_contains=keyword)
            print(f"Results: {results}")
        elif choice == "3":
            create_comprehensive_report(tracker)
        elif choice == "4":
            insights = generate_spending_insights(tracker)
            for i, insight in enumerate(insights, 1):
                print(f"{i}. {insight['message']}")
        elif choice == "5":
            create_comprehensive_report(tracker)
        elif choice == "6":
            exp_id = get_user_input_safe("Expense ID to edit: ")
            update_expense(tracker, exp_id, description="Edited")
        elif choice == "7":
            exp_id = get_user_input_safe("Expense ID to delete: ")
            delete_expense(tracker, exp_id)
        elif choice == "0":
            print("👋 Thank you for using the Finance Tracker!")
            break

        # Pause before showing menu again
        input("\nPress Enter to continue...")

# Run the interactive tracker
run_interactive_finance_tracker(my_finance_tracker)
````

### Step 6: Demonstrate Dictionary Mastery

Showcase advanced dictionary techniques:

````python
def demonstrate_dictionary_power(tracker):
    print("🎓 DEMONSTRATION OF DICTIONARY MASTERY")
    print("=" * 50)

    # 1. Dictionary Comprehension
    category_summary = {
        cat: {"spent": data["actual_spent"], "percent": (data["actual_spent"]/data["budget_amount"]*100) if data["budget_amount"]>0 else 0}
        for cat, data in tracker["categories"].items() if data["actual_spent"] > 0
    }
    print("Category summary:", category_summary)

    # 2. Nested navigation
    currency = tracker.get("metadata", {}).get("preferences", {}).get("currency", "EUR")
    print("Currency used:", currency)

    # 3. Data merging
    expense_analysis = {}
    for exp_id, exp in tracker["expenses"].items():
        cat = exp["category"]
        if cat not in expense_analysis:
            expense_analysis[cat] = {"total": 0, "count": 0}
        expense_analysis[cat]["total"] += exp["amount"]
        expense_analysis[cat]["count"] += 1
    print("Merged analysis:", expense_analysis)

    # 4. Advanced filtering
    filtered = {eid: exp for eid, exp in tracker["expenses"].items() if not exp["essential"] and exp["amount"] > 50}
    print("Non-essential expenses >50 EUR:", filtered)

    # 5. Aggregation by payment method
    payment_summary = {}
    for exp in tracker["expenses"].values():
        method = exp["payment_method"]
        if method not in payment_summary:
            payment_summary[method] = {"count": 0, "total": 0}
        payment_summary[method]["count"] += 1
        payment_summary[method]["total"] += exp["amount"]
    print("Summary by payment method:", payment_summary)


def create_learning_summary(tracker):
    print("\n📚 LEARNING SUMMARY")
    print("=" * 50)

    concepts = {
        "Dictionary creation": "Nested structures to organize data",
        "Safe access": "Using .get() to avoid errors",
        "Dictionary methods": "keys(), values(), items() for iteration and analysis",
        "Dynamic updates": "Adding, modifying, and deleting entries",
        "Complex filtering": "Comprehensions with multiple criteria",
        "Aggregation": "Grouping and statistical calculations"
    }

    for i, (concept, desc) in enumerate(concepts.items(), 1):
        print(f"{i}. {concept} → {desc}")

    print("\n📊 Project statistics:")
    print(f"   Tracked expenses: {tracker['analytics']['expense_count']}")
    print(f"   Managed categories: {len(tracker['categories'])}")
    print(f"   Main keys: {list(tracker.keys())}")

# Demonstration
demonstrate_dictionary_power(my_finance_tracker)
````
