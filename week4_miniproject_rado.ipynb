## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    # Validate expense amount
    try:
        amount = float(expense_input)
        if amount <= 0:
            print(" Please enter a positive number.")
            continue
    except ValueError:
        print(" Invalid input. Enter a number or 'done'.")
        continue

    # Get description
    description = input("What was this expense for? ")

    # Store values
    expenses.append(amount)
    descriptions.append(description)
    total_spent += amount

    # Show running total and budget percentage
    percent = (total_spent / monthly_budget) * 100
    print(f"✅ Added: ${amount:.2f} for {description}")
    print(f"📊 Running total: ${total_spent:.2f} ({percent:.1f}% of budget)")

print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    # Validate expense amount
    try:
        amount = float(expense_input)
        if amount <= 0:
            print(" Please enter a positive number.")
            continue
    except ValueError:
        print(" Invalid input. Enter a number or 'done'.")
        continue

    # Get description
    description = input("What was this expense for? ")

    # Store values
    expenses.append(amount)
    descriptions.append(description)
    total_spent += amount

    # Show running total and budget percentage
    percent = (total_spent / monthly_budget) * 100
    print(f"Added: ${amount:.2f} for {description}")
    print(f"Running total: ${total_spent:.2f} ({percent:.1f}% of budget)")

print(f"\nExpense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
# Expense Analysis Section
if len(expenses) > 0:
    print("\n Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # --- Find highest and lowest expenses (min/max pattern) ---
    highest_expense = expenses[0]
    lowest_expense = expenses[0]

    for amount in expenses:
        if amount > highest_expense:
            highest_expense = amount
        if amount < lowest_expense:
            lowest_expense = amount

    print(f"\n Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # --- Categorize expenses by size ---
    large_count = 0    # > 20% of budget
    medium_count = 0   # 5-20% of budget
    small_count = 0    # < 5% of budget

    for amount in expenses:
        percent = (amount / monthly_budget) * 100
        if percent > 20:
            large_count += 1
        elif percent >= 5:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # --- Smart Feedback ---
    print("\n Smart Feedback:")

    budget_used = (total_spent / monthly_budget) * 100

    if budget_used > 90:
        print(" Careful! You’ve used almost all of your budget.")
    elif budget_used > 50:
        print(" You’re halfway through your budget. Keep monitoring your spending.")
    else:
        print(" Great job! You’re managing your budget well so far.")

    if large_count > 2:
        print(" Consider reducing large expenses—they take up a big portion of your budget.")
    elif medium_count > large_count:
        print(" Most of your spending is moderate. Track these to avoid creeping over budget.")
    else:
        print(" Your spending is mostly small, which is easier to control.")

else:
    print("\n No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nEntering {num_goals} savings goals:")

# Collect information for each goal
for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Goal name
    name = input("Enter goal name: ")
    goal_names.append(name)

    # Target amount with validation
    target = -1
    while target <= 0:
        try:
            target = float(input("Enter target amount: "))
            if target <= 0:
                print("Target must be a positive number!")
        except ValueError:
            print("Please enter a valid number!")
    goal_targets.append(target)

    # Current savings with validation
    current = -1
    while current < 0 or current > target:
        try:
            current = float(input("Enter current amount saved: "))
            if current < 0:
                print("Amount cannot be negative!")
            elif current > target:
                print("Current amount cannot exceed target!")
        except ValueError:
            print("Please enter a valid number!")
    goal_current.append(current)

print(f"\nSaved {len(goal_names)} goals!")
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\nGoals Progress Analysis:")

    total_target = 0
    total_current = 0

    # Track progress to find weakest goal
    lowest_progress = 101
    weakest_goal_index = -1

    for i in range(len(goal_names)):
        name = goal_names[i]
        target = goal_targets[i]
        current = goal_current[i]

        # Progress percentage
        progress = (current / target) * 100 if target > 0 else 0

        # Amount remaining
        remaining = target - current

        # Months to completion (saving $100/month)
        monthly_saving = 100
        months = remaining / monthly_saving if remaining > 0 else 0

        print(f"\nGoal: {name}")
        print(f" - Target: ${target:.2f}")
        print(f" - Current: ${current:.2f}")
        print(f" - Progress: {progress:.1f}%")
        print(f" - Remaining: ${remaining:.2f}")
        print(f" - Estimated months to finish (saving $100/month): {months:.1f} months")

        # Track totals
        total_target += target
        total_current += current

        # Identify goal needing most attention
        if progress < lowest_progress:
            lowest_progress = progress
            weakest_goal_index = i

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\nOverall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Advice on weakest goal
    if weakest_goal_index != -1:
        weakest_name = goal_names[weakest_goal_index]
        weakest_progress = (goal_current[weakest_goal_index] / goal_targets[weakest_goal_index]) * 100

        print("\nGoal Needing the Most Attention:")
        print(f"'{weakest_name}' has the lowest progress at {weakest_progress:.1f}%.")

        # Custom encouragement message
        if weakest_progress < 20:
            print("Tip: Start with small weekly contributions to build momentum!")
        elif weakest_progress < 50:
            print("Tip: You're making progress! Try adding a bit more each month to accelerate.")
        else:
            print("Tip: You're more than halfway—keep going strong!")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Welcome to Your Finance Tracker Setup ===")

# -----------------------------
# Ask for name
# -----------------------------
name = input("Enter your name: ").strip()
if name == "":
    name = "User"
# -----------------------------
# Ask for monthly budget
# -----------------------------
while True:
    try:
        monthly_budget = float(input("\nEnter your monthly budget: $"))
        if monthly_budget <= 0:
            print("Budget must be positive!")
            continue
        break
    except ValueError:
        print("Please enter a valid number.")
# -----------------------------
# Ask for goals
# -----------------------------
goals = {}
print("\nNow let's set up your financial goals.")
print("Type 'done' as the goal name when finished.")

while True:
    goal_name = input("Goal name: ").strip()
    if goal_name.lower() == "done":
        break
    if goal_name == "":
        print("Goal name cannot be empty.")
        continue
    try:
        target = float(input("Target amount: $"))
        current = float(input("Current saved amount: $"))
        if target <= 0 or current < 0:
            print("Invalid amounts. Try again.")
            continue
        goals[goal_name] = {"current": current, "target": target}
    except ValueError:
        print("Enter valid numbers.")

# -----------------------------
# Ask for starting expenses
# -----------------------------
expenses = []
labels = []
print("\nLet's add your starting expenses.")
print("Enter expenses one by one. Type 'done' when finished.")

while True:
    entry = input("Expense amount (or 'done'): ").lower()
    if entry == "done":
        break
    try:
        amount = float(entry)
        if amount <= 0:
            print("Amount must be positive!")
            continue
        label = input("Label/description: ").strip()
        if label == "":
            label = "Misc"
        expenses.append(amount)
        labels.append(label)
    except ValueError:
        print("Please enter a valid number.")

print("\nSetup complete!Launching your finance tracker...")

# ============================================================
#                    MENU SYSTEM
# ============================================================

running = True

while running:
    print("\n" + "-"*30)
    print(f" {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f" Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    # ---------------------------------------------------
    # 1. Expense summary
    # ---------------------------------------------------
    if choice == "1":
        print("\n--- Expense Summary ---")

        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i, (amount, label) in enumerate(zip(expenses, labels), start=1):
                print(f"{i}. ${amount:.2f} - {label}")

            print("-" * 25)
            print(f"Total spent: ${sum(expenses):.2f}")

    # ---------------------------------------------------
    # 2. Goals progress
    # ---------------------------------------------------
    elif choice == "2":
        print("\n--- Goals Progress ---")

        if len(goals) == 0:
            print("No goals created yet.")
        else:
            for goal_name, data in goals.items():
                current = data["current"]
                target = data["target"]
                percent = (current / target) * 100 if target > 0 else 0
                print(f"{goal_name}: ${current}/${target} ({percent:.1f}% complete)")

    # ---------------------------------------------------
    # 3. Add new expense
    # ---------------------------------------------------
    elif choice == "3":
        print("\n--- Add New Expense ---")

        try:
            amount = float(input("Enter amount: $"))
            if amount <= 0:
                print("Amount must be positive!")
                continue

            label = input("Enter description (e.g., Coffee, Taxi): ").strip()
            if label == "":
                label = "Misc"

            expenses.append(amount)
            labels.append(label)

            print(f"Added: ${amount:.2f} - {label}")
            print("Expense added successfully!")

        except ValueError:
            print("Invalid number! Please enter a valid amount.")

    # ---------------------------------------------------
    # 4. Budget check
    # ---------------------------------------------------
    elif choice == "4":
        print("\n--- Budget Check ---")

        total_spent = sum(expenses)
        remaining = monthly_budget - total_spent
        used_percent = (total_spent / monthly_budget) * 100 if monthly_budget > 0 else 0

        print(f"Monthly budget: ${monthly_budget:.2f}")
        print(f"Total spent: ${total_spent:.2f}")
        print(f"Remaining: ${remaining:.2f}")
        print(f"Used: {used_percent:.1f}%")

        # Simple advice
        if used_percent < 40:
            print("You're doing great! Keep spending under control.")
        elif used_percent < 70:
            print("Moderate usage. Monitor your spending.")
        elif used_percent < 100:
            print("You're close to your budget limit. Slow down spending.")
        else:
            print("You've exceeded your monthly budget!")

    # ---------------------------------------------------
    # 5. Exit
    # ---------------------------------------------------
    elif choice == "5":
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
