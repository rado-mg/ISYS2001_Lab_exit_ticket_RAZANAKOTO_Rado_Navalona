## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### üí° Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
[write a python program that:

Allows users to input their monthly income
Allows users to input up to 3 expenses
Calculates remaining budget
Shows a warning if budget is negative]
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]
# Budget Calculator Program
# Tracks monthly income, up to 3 expenses, and calculates remaining budget

def main():
    print("=" * 50)
    print("       MONTHLY BUDGET CALCULATOR")
    print("=" * 50)

    # Get monthly income
    while True:
        try:
            monthly_income = float(input("\nEnter your monthly income: $"))
            if monthly_income < 0:
                print("Income cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    # Get expenses (up to 3)
    expenses = []
    print("\nEnter your expenses (up to 3). Press Enter to skip if you have fewer expenses.")

    for i in range(3):
        while True:
            try:
                expense_input = input(f"Expense {i + 1}: $")

                # Allow user to skip if they don't have 3 expenses
                if expense_input == "":
                    break

                expense = float(expense_input)
                if expense < 0:
                    print("Expense cannot be negative. Please try again.")
                    continue

                expenses.append(expense)
                break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        # If user skips, allow them to exit expense entry
        if expense_input == "" and len(expenses) > 0:
            skip_more = input("Skip remaining expenses? (yes/no): ").lower()
            if skip_more == "yes" or skip_more == "y":
                break

    # Calculate totals
    total_expenses = sum(expenses)
    remaining_budget = monthly_income - total_expenses

    # Display summary
    print("\n" + "=" * 50)
    print("           BUDGET SUMMARY")
    print("=" * 50)
    print(f"Monthly Income:      ${monthly_income:>10,.2f}")
    print("-" * 50)

    if expenses:
        for i, expense in enumerate(expenses, 1):
            print(f"Expense {i}:           ${expense:>10,.2f}")
        print("-" * 50)

    print(f"Total Expenses:      ${total_expenses:>10,.2f}")
    print("-" * 50)
    print(f"Remaining Budget:    ${remaining_budget:>10,.2f}")
    print("=" * 50)

    # Warning if budget is negative
    if remaining_budget < 0:
        print("\n  WARNING: Your budget is NEGATIVE!")
        print(f"   You are over budget by ${abs(remaining_budget):,.2f}")
    elif remaining_budget == 0:
        print("\n Your budget is balanced!")
    else:
        print(f"\n You have ${remaining_budget:,.2f} remaining in your budget.")

    print()

if __name__ == "__main__":
    main()
```

```python
# ENHANCED BUDGET TRACKER
# [Paste your enhanced code here]
# Budget Calculator Program
# Tracks monthly income, up to 3 expenses, calculates remaining budget, and shows percentage spent

def main():
    print("=" * 50)
    print("       MONTHLY BUDGET CALCULATOR")
    print("=" * 50)

    # Get monthly income
    while True:
        try:
            monthly_income = float(input("\nEnter your monthly income: $"))
            if monthly_income < 0:
                print("Income cannot be negative. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    # Get expenses (up to 3)
    expenses = []
    print("\nEnter your expenses (up to 3). Press Enter to skip if you have fewer expenses.")

    for i in range(3):
        while True:
            try:
                expense_input = input(f"Expense {i + 1}: $")

                # Allow user to skip if they don't have 3 expenses
                if expense_input == "":
                    break

                expense = float(expense_input)
                if expense < 0:
                    print("Expense cannot be negative. Please try again.")
                    continue

                expenses.append(expense)
                break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

        # If user skips, allow them to exit expense entry
        if expense_input == "" and len(expenses) > 0:
            skip_more = input("Skip remaining expenses? (yes/no): ").lower()
            if skip_more == "yes" or skip_more == "y":
                break

    # Calculate totals
    total_expenses = sum(expenses)
    remaining_budget = monthly_income - total_expenses

    # Calculate percentage of income spent
    if monthly_income > 0:
        percentage_spent = (total_expenses / monthly_income) * 100
        percentage_remaining = (remaining_budget / monthly_income) * 100
    else:
        percentage_spent = 0
        percentage_remaining = 0

    # Display summary
    print("\n" + "=" * 50)
    print("           BUDGET SUMMARY")
    print("=" * 50)
    print(f"Monthly Income:      ${monthly_income:>10,.2f}")
    print("-" * 50)

    if expenses:
        for i, expense in enumerate(expenses, 1):
            print(f"Expense {i}:           ${expense:>10,.2f}")
        print("-" * 50)

    print(f"Total Expenses:      ${total_expenses:>10,.2f}")
    print(f"Percentage Spent:    {percentage_spent:>10.1f}%")
    print("-" * 50)
    print(f"Remaining Budget:    ${remaining_budget:>10,.2f}")
    print(f"Percentage Remaining:{percentage_remaining:>10.1f}%")
    print("=" * 50)

    # Warning if budget is negative
    if remaining_budget < 0:
        print("\n  WARNING: Your budget is NEGATIVE!")
        print(f"   You are over budget by ${abs(remaining_budget):,.2f}")
        print(f"   You spent {percentage_spent:.1f}% of your income!")
    elif remaining_budget == 0:
        print("\n Your budget is balanced!")
        print(f"   You spent 100% of your income.")
    else:
        print(f"\n You have ${remaining_budget:,.2f} remaining in your budget.")
        print(f"   You spent {percentage_spent:.1f}% of your income.")

    print()

if __name__ == "__main__":
    main()
```
